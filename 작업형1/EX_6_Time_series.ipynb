## 2번

## 4번 
Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

### `pd.DateOffset` 객체
- pd.DateOffset 객체를 사용하여 날짜 및 시간에 일정한 간격을 더하거나 빼는 작업을 수행할 수 있음

```python
df.loc[df['Yr_Mo_Dy'].dt.year >= 2061, 'Yr_Mo_Dy'] -= pd.DateOffset(years=100)
df.head()
```

## 7번
weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼을 만들어라

```python
df["WeekCheck"] = df["weekday"].map(lambda x : 1 if x in [5,6] else 0)
df.head()
```

## 10번
년도 - 월을 기준으로 모든 컬럼의 평균값을 구하여라

```python
df.groupby([df["Yr_Mo_Dy"].dt.year, df["Yr_Mo_Dy"].dt.month]).mean()
```

### `dt.to_period("M")`
- dt.to_period('M')은 날짜를 월(Month) 단위의 기간으로 변환하는 메서드 
- 'M'은 월 단위를 의미 
- 'D'는 일 단위, 'Q'는 분기(Quarter) 단위를 의미 

$\rightarrow$ 이렇게 변환된 기간은 groupby에서 사용

```python
Ans = df.groupby(df["Yr_Mo_Dy"].dt.to_period('M')).mean()
Ans
```

## 12번
RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하여라

```python
ma_rpt = df["RPT"].rolling(window=7).mean()
ma_val = df["VAL"].rolling(window=7).mean()

print(ma_rpt)
print(ma_val)
```
