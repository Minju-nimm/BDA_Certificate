````python
import pandas as pd

channel = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
````

## 1번
각 데이터의 ‘ct’컬럼을 시간으로 인식할수 있게 datatype을 변경하고 video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라

````python
channel["ct"] = pd.to_datetime(channel["ct"])
video["ct"] = pd.to_datetime(video["ct"])

result = video["videoname"].value_counts()
print(result)
````

## 2번
수집된 각 video의 가장 최신화 된 날짜의 viewcount(viewcnt)값을 출력하라

- sort_values로 내림차순 적용하는 아이디어 고려
- 수집된 '각' video 임을 고려

### `drop_duplicates(, keep="last")`
- videoname 열을 기준으로 중복된 행을 제거
    -  여러 개의 중복된 행이 있을 경우, 마지막 행만 유지
- videoname 값이 동일한 행들 중에서 가장 마지막에 나오는 행만 유지

````python
# videoname과 ct를 기준으로 정렬하고, videoname을 기준으로 하여 최신의 값만 유지하고 나머지는 중복행 제거
sort_df = video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last').reset_index(drop=True)
sort_df[["viewcnt", "videoname", "ct"]]
````

## 4번
각 채널의 2021-10-03 03:00:00, 2021-11-01 15:00:00 구독자수(subcnt) 의 증가량(차이)를 구하여라

### dt.strftime( )을 통해 변환 후 조건 설정

````python
channel = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
channel.head()
````

````python
channel["ct"] = pd.to_datetime(channel["ct"], format = "%Y-%m-%d %H")
channel.info()
````

````python
# dt.strftime을 통해 다음과 같은 형식으로 변환
end = channel[channel["ct"].dt.strftime('%Y-%m-%d %H') =='2021-11-01 15']
start = channel[channel["ct"].dt.strftime('%Y-%m-%d %H') =='2021-10-03 03']

display(end.head())

end_df = end[['channelname','subcnt']].reset_index(drop=True)
start_df = start[['channelname','subcnt']].reset_index(drop=True)

end_df.columns = ['channelname','end_sub']
start_df.columns = ['channelname','start_sub']

display(end_df.head())
display(start_df.head())
````

````python
total = pd.merge(start_df, end_df)
total

total['dif'] = total['end_sub'] - total['start_sub']
result = total[['channelname','dif']]
display(result)
````

### ct 열의 값들 간의 차이를 계산, diff(1) 사용
- 현재 행과 이전 행 간의 차이를 계산하는 메서드입니다.

### timedelta : 시간 간격을 나타내는 timedelta 객체를 생성

````python
df_ct = video[video["videoname"].str.contains("1")].sort_values("ct").reset_index(drop=True)
# df_ct

# 수집간격 5분 이하, 20분 이상인 데이터 구간의 시각 모두 출력
import datetime

# ct칼럼에서 이전행과의 차이를 기준으로 시간 간격 계산 
df_ct[
    (df_ct["ct"].diff(1) >= datetime.timedelta(minutes =20)) |  (df_ct["ct"].diff(1) <= datetime.timedelta(minutes = 5))
    ]
````

````python
# 720, 722, 1636을 기준으로 오른쪽 숫자도 포함
answer = df_ct[df_ct.index.isin([720,721,722,723,1635,1636,1637])]
display(answer)
````

## 8번
video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

````python
video.head()
````

````python
# 각 에피소드의 가장 '최근' 데이터
last_ep = video.sort_values("ct").drop_duplicates("videoname", keep="last")
# last_ep

last_ep["ratio"] = last_ep["dislikecnt"] / last_ep["likecnt"]

last_ep.sort_values("ratio")[["videoname", "ratio"]].reset_index(drop=True)
````

## 9번
2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오
